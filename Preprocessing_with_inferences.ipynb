##### Statistical analysis

```python
ratings_per_joke={}
r1=pd.read_excel('unzipped_files/jester-data-1.xls',header=None)
r2=pd.read_excel('unzipped_files/jester-data-2.xls',header=None)
r3=pd.read_excel('unzipped_files/jester-data-3.xls',header=None)
l1=range(len(r1[0]))
l2=range(len(r2[0]))
l3=range(len(r3[0]))
for i in range(1,101):
    c=0
    for j in l1:
        if(r1[i][j]==99):
            r1[i][j]=np.nan
        else:
            c+=1
    for j in l2:
        if(r2[i][j]==99):
            r2[i][j]=np.nan
        else:
            c+=1
    for j in l3:
        if(r3[i][j]==99):
            r3[i][j]=np.nan
        else:
            c+=1
    ratings_per_joke['joke_'+str(i)]=c
final_data=pd.concat([r1,r2,r3])
final_data.columns=['no_of_ratings']+['joke_'+str(i+1) for i in range(100)]
# final_data.to_csv('jester_ratings_preprocessed1.csv', index=False)
display(final_data)
```

##### Data Cleaning

```python
for i in range(1,101):
    j=(description['no_of_ratings'][1] -description['joke_'+str(i)][1])*100/description['no_of_ratings'][1]
    if j>=75:
        print("Too Much NaNs in joke_"+str(i)+" % = "+str(j))
    elif j<=10:
        print("Very LESS NaNs in joke_"+str(i)+" % = "+str(j))
```

None of Columns have very high no. of Nans so we donot need to drop any joke (All jokes have been rated by atleast 25% users of our dataset)
But ther are certain jokes which have very less amount of missing values.
To clean these coloumns we have two options-> 
1) To either drop rows wich have null values for this coloumns
2) To fill these values with some value
Since in this dataset values in colomns may depand on different rows we will refrain ou selves from doing the first one

Policy for Data cleaning -> we would be replacing all the null values with a random value from gaussian distribution(jokes'mean,jokes'std) in range of -10 to 10
{I think This is a just policy because this will try to retain the earlier distribution of dataset }
Other policies can be
1) values with a random value between-1+joke's mean 1+joke's mean in range of -10 to 10
2) mean
3) median
4) 0 (To represent equal chances of bad and good joke)

```python
print("Corelation_matrix")
corelation_matrix=covarience_matrix.copy()
for i in range(covarience_matrix.shape[0]):
    for j in range(covarience_matrix.shape[0]):
        corelation_matrix[i][j]=covarience_matrix[i][j]/(description[2][i]*description[2][j])
print("Index are the column-index of dataset")
print(normalized_data.columns)
display(pd.DataFrame(corelation_matrix))
```

```python
print("Correlation between different user inputs (only jokes)")
def generate_corelationrow(i,n1=normalized_data.to_numpy()):
    list_corr_i=[]
    for j in l1:
        cv=0
        vi=0
        vj=0
        for k in range(2,102):
            cv+=n1[i][k]*n1[j][k]
            vi+=n1[i][k]*n1[i][k]
            vj+=n1[j][k]*n1[j][k]
        list_corr_i.append(cv/((vi**0.5)+(vj**0.5)))
    return list_corr_i
print("Dataset of corelation is preaty huge!! It might take 660 hours to generate it")
print("Thus we are here generating only one corelation row (as an example) for user_id 2") 
display(pd.DataFrame(generate_corelationrow(2)).T)
    
    
    
```
